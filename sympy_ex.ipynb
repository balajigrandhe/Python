```python
from sympy import symbols, Symbol, factor, expand,solve
```

```python
a,b = symbols('a,b')
expand_exp = expand((a+b)**3)
```

```python
print(expand_exp)
```

```python
fact = factor((a+b)**3)
print(fact)
```

```python
from sympy.plotting import plot
x,y=symbols('x,y')
%matplotlib inline
plot(x+4,3*x)
solve((x+4-y,3*x-y),dict=True)
```
